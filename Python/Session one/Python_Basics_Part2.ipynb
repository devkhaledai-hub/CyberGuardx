## Numbers Examples

```python
# Integer examples
a = 10
b = -5
c = 0
print(a, b, c)
```

```python
# Float examples
x = 3.14
y = -0.5
z = 2.0
print(x, y, z)
```

```python
# Arithmetic operations
a = 10
b = 3
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

```python
# Built-in number functions
x = -10
print("abs:", abs(x))
print("pow:", pow(2, 4))
print("round:", round(3.6))
print("max:", max(1, 5, 3))
print("min:", min(1, 5, 3))
```

```python
# Type conversion
x = 10
print(float(x))

y = 3.7
print(int(y))

z = "100"
print(int(z))
```

## Strings Examples

```python
# Creating strings
name = "Python"
message = 'Hello World'
multi = """This is
multi line
string"""

print(name)
print(message)
print(multi)
```

```python
# String indexing
text = "Python"
print(text[0])
print(text[1])
print(text[-1])
print(text[-2])
```

```python
# String slicing
text = "Python Programming"
print(text[0:6])
print(text[7:18])
print(text[:6])
print(text[7:])
print(text[::-1])
```

```python
# String step slicing
text = "ABCDEFGHIJ"
print(text[::2])
print(text[::3])
print(text[::-1])
```

```python
# String operations
a = "Hello"
b = "Python"
print(a + " " + b)
print(a * 3)
print("H" in a)
print("z" in a)
```

```python
# String methods
text = " hello python "
print(text.upper())
print(text.lower())
print(text.strip())
print(text.replace("python", "world"))
print(text.find("python"))
```

```python
# Exercise
# Reverse a string
```

```python
# Solution
text = "Python"
print(text[::-1])
```

## Boolean Examples

```python
# Boolean values
a = True
b = False
print(a, b)
```

```python
# Comparison operators
print(10 > 5)
print(10 < 5)
print(10 == 10)
print(10 != 5)
```

```python
# Boolean with variables
x = 10
y = 20
print(x > y)
print(x < y)
```

```python
# Logical operators
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

```python
# Boolean from expressions
print(bool(1))
print(bool(0))
print(bool("Hello"))
print(bool(""))
```
