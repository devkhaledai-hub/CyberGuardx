## Data Types

```python
# Integer
x = 10
print(x, type(x))
```

```python
# Float
x = 3.14
print(x, type(x))
```

```python
# String
x = "Python"
print(x, type(x))
```

```python
# Bytes
x = bytes([65, 66, 67])
print(x, type(x))
```

```python
# Boolean
x = True
print(x, type(x))
```

```python
# List
x = [1, 2, 3]
print(x, type(x))
```

```python
# Tuple
x = (1, 2, 3)
print(x, type(x))
```

```python
# Dictionary
x = {"name": "Ali", "age": 25}
print(x, type(x))
```

```python
# Set
x = {1, 2, 3}
print(x, type(x))
```
